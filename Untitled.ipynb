```python
def typeCompression(df):
    for col in df.columns:
        if(df[col].dtype == 'float64'):
            df[col] = df[col].astype('float32')
    return df

def sparsityReducer(df):
    user_df = df.groupby('User')['Rating'].agg(['count','mean'])
    item_df = df.groupby('Item')['Rating'].agg(['count','mean'])

#     print(user_df)

    user_threshold = user_df['count'].quantile(0.2)
    item_threshold = item_df['count'].quantile(0.2)

    drop_user = user_df[user_df['count']<user_threshold].index
    drop_item = item_df[item_df['count']<item_threshold].index

#     print(drop_user,drop_item)
    
    print('Original Shape: {}'.format(df.shape))
    df = df[~df['Item'].isin(drop_item)]
    df = df[~df['User'].isin(drop_user)]
    print('After Trim Shape: {}'.format(df.shape))
    
    return df

def sparseChanger(df,min_r, max_r):
    user_df = df.groupby('User')['Rating'].agg(['count','mean'])
    #print(user_df)
    #user_threshold = user_df['count'].quantile(0.2)
    
#     if(threshold==50):
    drop_user = user_df[(user_df['count']>min_r) & (user_df['count']<max_r)].index
#     else:
#         drop_user = user_df[user_df['count']>800].index
    #print(drop_user)
    
    print('Original Shape: {}'.format(df.shape))
    df = df[~df['User'].isin(drop_user)]
    print('After Trim Shape: {}'.format(df.shape))
    
    return df

def generateConcepts(filename, nrows, ncols):
    #c = concepts.load_csv(filename)
    df = pd.read_csv(filename, index_col='User', nrows = nrows,usecols = np.arange(0,ncols))
#     df = pd.read_csv(filename, index_col='User', nrows = nrows)
    
    max_val = int(max(df.columns))
    #objects = df.index.tolist()
    objects = [x for x in range(max_val+1, max_val+df.shape[0]+1)]

    properties = list(df.columns)

    bools = list(df.fillna(False).astype(bool).itertuples(index=False, name=None))

    c_ = concepts.Context(objects, properties, bools)
    
    lattice = c_.todict()['lattice']
    
    final_concept=[None] * len(lattice)
    
#     print(c_.todict())
    
#     return list(map(lambda v: [v[0],v[1]], lattice))
#     return [[c1, c2] for c1, c2, c3, c4 in lattice]
#     return list(filter(lambda x: [x[0],x[1]], chain(*zip_longest(*lattice))))
#     return lattice
#     print('here5')
    
#     for index, row in lattice_df.iterrows():
#          final_concept[index] = [row[0],[row[1]]]

    for i, c in enumerate(lattice):
        final_concept[i] = [c[0],[c[1]]]
    
    return final_concept
```

```python
class GeneticEvolution:
    
    def __init__(self,initial_population,target_user, mutation_probability_type='rpm'):
        self.initial_population = initial_population
        self.target_user = target_user
        self.fittest_user = None
        self.mutation_probability_type = mutation_probability_type
        
    def initiate(self):
        generation = 1
        while len(self.target_user) != self.get_row_value_count(self.target_user):
            self.selection()
            self.crossover()
            self.mutation()
            
            print("Geneartion {}: fitness: {}".format(generation, self.get_row_value_count(self.target_user)))
            
            generation += 1
        
        return self.target_user, generation
    
    def get_row_value_count(self, row):
        return row.apply(lambda v: v != 0 and not pd.isna(v)).sum()
        
    def selection(self):
        count = 0
        for index, row in self.initial_population.iterrows():
            row_count = self.get_row_value_count(row)
            if count < row_count:
                self.fittest_user = row
                count = row_count
    
    def crossover(self):
        crossover_range = int(len(self.target_user)*0.1)
        start_point = random.randint(0, len(self.target_user) - crossover_range)
        end_point = start_point + crossover_range
        for fittest, target in zip(self.fittest_user[start_point:end_point], self.target_user[start_point:end_point]):
            if fittest == 0 and target == 0:
                fittest = self.target_user.mean()
                target = self.fittest_user.mean()
            else:
                temp = fittest
                fittest = target
                target = temp
        
    def mutation(self):
        if self.mutation_probability_type == 'rpm':
            mutation_point = random.choice(self.initial_population.columns)
        elif self.mutation_probability_type == 'spm':
            if random.randint(1, 10) <= int(self.mutation_probability/10):
                
#         mutation_point = random.randint(0, len(self.target_user)-1)
        
        if self.fittest_user[mutation_point] == 0:
            self.target_user[mutation_point] = self.fittest_user.mean()
        else:
            self.target_user[mutation_point] = self.fittest_user[mutation_point]
    
```

```python
def get_population_unique_rows(final_concept, user):
    init_population_rows=[]
    for concept in final_concept:
        if user in concept[0] and len(concept[0]) < 200:
            init_population_rows.extend(concept[0])
            
    init_population_rows = list(set(init_population_rows))
    
    return init_population_rows
```

```python
def get_initial_population(df, rows):
    return df.loc[rows].fillna(0)
```
